# Instance methods

```python
class Student:
    def __init__(self, name, age):
            self.name = name
            self.age = age
            self.grade = None
    
    # instance method
    def description(self):
        return "Student {} is {} years old".format(self.name, self.age)
    
    def get_grade(self):
        return self.grade if self.grade else "No grade yet"
    
    def __str__(self):
        return "({},{},{})".format(self.name, self.age, self.grade)
```

```python
new_student = Student("Thomas", 23)
another_student = Student("Stephany", 25)
print(new_student.__str__())
print(another_student)
```

```python
class Student:
    def __init__(self, name, age):
            self.name = name
            self.age = age
            self.grade = None
    
    # instance method
    def description(self):
        return "Student {} is {} years old with grade: {}".format(self.name, self.age, str(self.grade))
    
    def get_grade(self):
        return self.grade if self.grade else "no grade yet"
    
    def study_hours(self, hours):
        if hours >= 15: 
            self.grade = 'A'
        elif 20 > hours >= 15:
            self.grade = 'B'
        elif 15 > hours >= 10:
            self.grade ='C'
        else:
            self.grade = 'D'
            
    def __str__(self):
        return "({},{},{})".format(self.name, self.age, self.grade)
```

```python
new_student = Student("Chen", 23)
```

```python
new_student.description()
```

```python
new_student.study_hours(16)
```

```python
objviz(new_student)
```

```python
new_student.description()
```

```python
new_student.study_hours(20)
```

```python
objviz(new_student)
```

```python
new_student.description()
```

```python
class Human:
    def __init__(self, age, name):
        self.name = name
        self.age = age
        
    def __str__(self):
        return "HUMAN: (%s, %s)" %(self.name, self.age)
    
class Student(Human):
    def __init__(self, name, age, grade = None):
            super().__init__(name, age)
            self.grade = None
    
    # instance method
    def description(self):
        return "Student {} is {} years old with grade: {}".format(self.name, self.age, str(self.grade))
    
    def get_grade(self):
        return self.grade if self.grade else "no grade yet"
    
    def study_hours(self, hours):
        if hours >= 20: 
            self.grade = 'A'
        elif 20 > hours >= 15:
            self.grade = 'B'
        elif 15 > hours >= 10:
            self.grade ='C'
        else:
            self.grade = 'D'
            
    def __str__(self):
        return super().__str__() + " is student with grade %s" %self.grade
```

```python
new_student = Student("Chistine", 23)
print(new_student)
```

```python
objviz(new_student)
```
