# Brute force
The brute-force algorithm is a “loop join”, i.e., traversing through all the elements of one array and comparing them to the elements of the other array. Let m and n be the lengths of the two input arrays.

```python
A= [2, 3, 3, 5, 5, 6, 7, 7, 8, 12]
B= [5, 5, 6, 8, 8, 9, 10, 10]


def intersect_two_sorted_arrays(A, B):
    return [elem for indx, elem in enumerate(A) if (indx == 0 or elem != A[indx - 1]) and elem in B]

print(intersect_two_sorted_arrays(A,B))











set_answer = set(A).intersection(set(B))
list_answer = list(set_answer)
list_answer.sort()
print(list_answer)
```

```python
import bisect
"""
This module provides support for maintaining a list in sorted order without having to sort the list 
after each insertion. 
The bisect() function is generally useful for categorizing numeric data.
"""
def intersect_two_sorted_arrays(A, B): 
    def is_present(k):
        indx = bisect.bisect_left(B, k) 
        return indx < len(B) and B[indx] == k
    
    return [
        elem for indx, elem in enumerate(A)
        if (indx == 0 or elem != A[indx - 1]) and is_present(elem) 
    ]
print(intersect_two_sorted_arrays(A,B))
```

```python
A= [2, 3, 3, 5, 5, 6, 7, 7, 8, 12]
B= [5, 5, 6, 8, 8, 9, 10, 10]

def intersect_two_sorted_arrays(A, B):
    indx_a = 0
    indx_b = 0 
    intersection_A_B = [] 
    while indx_a < len(A) and indx_b < len(B):
        if A[indx_a] == B[indx_b]:
            if indx_a == 0 or A[indx_a] != A[indx_a - 1]: 
                intersection_A_B.append(A[indx_a])
            indx_a += 1
            indx_b += 1
        elif A[indx_a] < B[indx_b]: 
            indx_a += 1
        else: # A[indx_a] > B[indx_b].
            indx_b += 1
    return intersection_A_B

print(intersect_two_sorted_arrays(A, B))
```
